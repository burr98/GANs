```python
#import libraries
import pandas as pd
from tgan.model import TGANModel
```

```python
df = pd.read_csv('/content/advertising.csv')

df.sample(10)
```

```python
#show the number of unique values in all the columns in the Dataframe
df.nunique().sort_values(ascending=False)
```

```python
df.info()
```

```python
#find the number of null values in the Dataframe

df.isnull().sum()
```

```python
df.skew()
```

```python
#rename the Male column to gender
df = df.rename({'Male':'Gender'}, axis = 1)

df.sample(10)
```

```python
#rename the values in the Gender column
df = df.replace({'Gender' : {1 : 'Male', 0:'Female'}})

df.sample(3)
```

```python
cols = df.columns.tolist()
```

```python
tgan_ = TGANModel(cols, max_epoch=3, batch_size=4)
```

```python
tgan_.fit(df)
```

```python
samples = tgan_.sample(1000)

samples.head()
```

```python
samples.shape
```

```python
samples['Timestamp'] = pd.to_datetime(samples['Timestamp'])

samples.info()
```

```python
num_cols = ['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage']

samples[num_cols] = samples[num_cols].astype(str).astype('float')
samples.describe()
```
